```python
# Function to plot ROC and AUC
def plot_roc_auc(y_test, y_prob, target_name):
    y_test_bin = label_binarize(y_test, classes=[0, 1])
    fpr, tpr, _ = roc_curve(y_test_bin, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic - {target_name}')
    plt.legend(loc="lower right")
    plt.show()


y_prob_damper = multi_target_rf.estimators_[0].predict_proba(X_test)[:, 1]
y_prob_inclination = multi_target_rf.estimators_[1].predict_proba(X_test)[:, 1]
# Plot ROC and AUC for Damper
plot_roc_auc(y_test['Damper'], y_prob_damper, "Damper")

# Plot ROC and AUC for Inclination
plot_roc_auc(y_test['Inclination'], y_prob_inclination, "Inclination")
```
